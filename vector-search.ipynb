```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
# from langchain_openai import OpenAIEmbeddings
# from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.embeddings import JinaEmbeddings
from langchain_chroma import Chroma
```

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
import pandas as pd

books = pd.read_csv("books_cleaned.csv")
```

```python
books
```

```python
books["tagged_description"].to_csv("tagged_description.txt",
                                   sep="\n",
                                   index=False,
                                   header=False)
```

```python
raw_documents = TextLoader("tagged_description.txt", encoding='utf-8').load()
text_splitter = CharacterTextSplitter(
    # chunk_size=1,
    chunk_size=1000,
    # chunk_overlap=0,
    chunk_overlap=100,
    separator="\n"
)
documents = text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
# embedding = GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001")
embedding = JinaEmbeddings(model="jina-embeddings-v2-base-en")

# db_books = Chroma.from_documents(
#     documents,
#     embedding=embedding)
db_books = None
batch_size = 512  # Jina API limit

for i in range(0, len(documents), batch_size):
    batch = documents[i : i + batch_size]
    if db_books is None:
        db_books = Chroma.from_documents(batch, embedding=embedding)
    else:
        db_books.add_documents(batch)
    print(f"✅ Embedded batch {i//batch_size + 1} ({len(batch)} docs)")

print(f"\n🎉 Finished creating vector DB with {len(documents)} chunks!")
```

```python
query = "A book about to teach children about nature"
docs = db_books.similarity_search(query, k=10)
docs
```

```python
books[books["isbn13"] == int(docs[0].page_content.split()[0].strip())]
```

```python
def retrieve_semantic_recommendations(
        query: str,
        top_k: int = 10,
) -> pd.DataFrame:
    recs = db_books.similarity_search(query, k = 50)

    books_list = []

    for i in range(len(recs)):
        books_list += [int(recs[i].page_content.strip('"').split()[0])]

    return books[books["isbn13"].isin(books_list)].head(top_k)
```

```python
retrieve_semantic_recommendations(query)
```
